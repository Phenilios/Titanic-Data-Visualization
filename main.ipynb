```python
line_fare = df[(df['Ticket'] == 'LINE') & (df['Fare'] > 0)]
line_fare
```

```python
df = df[df['Fare'] > 0].copy()
df.reset_index(drop=True, inplace=True)
```

```python
df.info()
```

```python
df.info()
```

```python
# data visualization
plt.hist(df['Age'], bins=10, edgecolor='black', color='lightgreen')
plt.xlabel("Age")
plt.ylabel("Count")
plt.title("Age Distribution")
plt.show()
```

```python
df.groupby('Sex')['Survived'].mean().plot(
    kind='pie',
    autopct='%1.2f%%',
    shadow=True,
    colors=['pink', 'lightblue'],
    wedgeprops={'edgecolor': 'black'}
)

plt.ylabel('')
plt.title('Survival Rate by Sex')
plt.show()
```

```python
fare_survival_total = df.groupby('Survived')['Fare'].sum()
fare_survival_total.plot(kind='barh', color=['red', 'green'])
plt.yticks([0, 1], ['No', 'Yes'])
plt.xlabel('Total Fare')
plt.ylabel('Survived')
plt.title('Total Fare by Survival')
plt.show()
```
